## ✅ Read Modes

Read modes define how Spark should behave when it reads a file/table and encounters bad, corrupt, or unexpected data.They are set using:

###.option("mode", "PERMISSIVE")

###1)PERMISSIVE(Default):
Replace all invalid values with Null
✅ Loads all valid rows
✅ Corrupt rows go into a special column called _corrupt_record
✅ Job does NOT fail

###2)FAILFAST:
❌ If Spark finds even 1 bad row → it stops the entire job
✅ Good for strict pipelines / financial data
✅ Ensures data integrity

###3)DROPMALFORMED:
✅ Drops (ignores) rows that don’t match the schema
✅ Keeps only the correct rows
✅ No error thrown

### ✅ Write Modes

###1)overwrite
Deletes existing data and writes new data	Rebuilding a table or reprocessing clean data

###2)append	
Adds new rows to existing data	Incremental loads

###3)ignore	
Does nothing if data exists	Safe writes where duplicates should be skipped

###4)errorifexists (default)	
Throws error if data exists	Protect data integrity.

🔹 What Is MERGE?

MERGE is NOT a normal write mode.
It is a feature of Delta Lake used for upserts.
MERGE is used for:
Upserts (update + insert)
Slowly Changing Dimensions
CDC (change data capture)
Syncing tables

It works only with Delta tables.

🔹 How MERGE Works in Spark (Delta)

Example:

from delta.tables import DeltaTable

deltaTable = DeltaTable.forPath(spark, "/mnt/sales")

deltaTable.alias("target").merge(
    source_df.alias("source"),
    "target.id = source.id"
).whenMatchedUpdateAll() \
 .whenNotMatchedInsertAll() \
 .execute()

#Enforcing Schema
Schema can be created using 2 types
1. Python
2. SQL DDL

### Python
from pyspark.sql.types import StructType,StructField

StructType- List of StructField (defines structure of dataframe)

StructField- Details about fields (columns)

**my_schema=StructType([StructField(Name,Data_type,Nullable)])**

**.schema(my_schema)**

###Schema cannnot be applied to an already loaded DataFrame. Schema must be applied before loading the data or during reading the file from volumes


```python
people_path ="/Volumes/workspace/default/workspace_volume/sample_databricks.csv"
df_people = spark.read.option("header", True)\
.option("inferSchema", True)\
.csv(people_path)
df_people.printSchema()
df_people.show(1)
```

```python
from pyspark.sql.types import StructType, StructField, IntegerType, StringType

my_schema=StructType([StructField("Id",IntegerType(),True),\
                    StructField("Person Name",StringType(),True),\
                    StructField("Person age",IntegerType(),True),\
                    StructField("Birth City",StringType(),True),\
                    StructField("Pension",StringType(),True)])

df_people = spark.read\
    .option("header", True) \
    .schema(my_schema) \
    .csv(people_path)

df_people.printSchema()

df_people.show(5)
```
